## 1. (Graph) What was the runtime for each movie?

````python
#https://pandas.pydata.org/pandas-docs/version/0.23.1/generated/pandas.DataFrame.plot.bar.html
#https://stackoverflow.com/questions/25447700/annotate-bars-with-values-on-pandas-bar-plots
#https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar_label.html#matplotlib.axes.Axes.bar_label

ax = movies_data[['movie_title', 'runtime']].plot(kind='barh', x='movie_title', rot=45, legend=False, figsize=(10, 7))
for container in ax.containers:
    ax.bar_label(container)
plt.title("Movie Runtimes")
plt.xlabel("Runtime in minutes")
plt.ylabel('')
````

### The above graph shows the runtimes for each of the movies from the original Oscar winners CSV. The graph was generated by plotting the 'movie_title' and 'runtime' values from the 'movie_data' CSV using a barh plot. Using barh, in addition to rotating the movie name values, allows for a better fit on screen. 

## 2. What movie had the longest runtime?

````python
longest_movie = movies_data[['movie_title','runtime']].nlargest(n=3, columns='runtime')
print(longest_movie)
````

### At 134 minutes, '12 Years a Slave' is the movie with the longest runtime. This was determined by using ```.nlargest``` to sort the runtimes in the 'movies_data' file and returning the three largest runtimes.

## 3. Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?

````python
#https://stackoverflow.com/questions/36420908/can-i-draw-a-regression-line-and-show-parameters-using-scatterplot-with-a-pandas
#https://www.geeksforgeeks.org/pandas-scatter-plot-dataframe-plot-scatter/
#https://www.statology.org/seaborn-figure-size/

runtime_x_awards_won = sns.lmplot(x='runtime', y='award_wins', data=movies_data, fit_reg=True, height=6, aspect=1.5)
plt.show(runtime_x_awards_won)
````

````python
#https://datascienceparichay.com/article/get-correlation-between-columns-of-pandas-dataframe/
#https://sphweb.bumc.bu.edu/otlt/MPH-Modules/PH717-QuantCore/PH717-Module9-Correlation-Regression/PH717-Module9-Correlation-Regression4.html

# correlation between Col1 and Col2
corr_runtime_awards = movies_data['runtime'].corr(movies_data['award_wins'])
print(corr_runtime_awards)
````

### With a correlation coefficient of 0.11, the relationship between the length of the movie and the number of awards a movie won would be considered nonexistent or very weak. The graph to show the extent of the relationship was generated using Seaborn ```lmplot```, which shows the the regression line for the given data points. To determine the correlation value, ```.corr``` was used to compare the values in the 'runtime' and 'award_wins' columns. 

## 4. (Graph) How many awards and nominations did each movie receive?

````python
ax = movies_data[['movie_title', 'award_wins', 'award_nominations']].plot(kind='barh', x='movie_title', rot=45, figsize=(10, 7))
for container in ax.containers:
    ax.bar_label(container)
    
plt.title("Award Wins & Nominations")
plt.ylabel('')
````

### To see the number of award wins and nominations, a horizontal bar graph was plotted with the value of the 'award_wins' and 'award_nominations' columns were plotted against the 'movie_title' column. Additionally, the values for each bar were added to improve readability.
